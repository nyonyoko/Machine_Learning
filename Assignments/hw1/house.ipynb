```python
# Import the libraries
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import statsmodels.api as sm
```

```python
# Read the CSV file and set the first row as the header
df = pd.read_csv('housingUnits.csv', header=0)
```

# Question 1

## Standardizing 2 and 3 by themselves

```python
# Standardize the total_rooms and total_bedrooms variables
df['standardized_total_rooms'] = stats.zscore(df['total_rooms'])
df['standardized_total_bedrooms'] = stats.zscore(df['total_bedrooms'])
```

```python
# Histograms
plt.figure(figsize=(16, 10))

plt.subplot(2, 2, 1)
plt.hist(df['total_rooms'], bins=50, alpha=0.5, label='Original')
plt.xlabel('Total Rooms')
plt.legend()

plt.subplot(2, 2, 2)
plt.hist(df['standardized_total_rooms'], bins=50, alpha=0.5, label='Standardized')
plt.xlabel('Standardized Total Rooms')
plt.legend()

plt.subplot(2, 2, 3)
plt.hist(df['total_bedrooms'], bins=50, alpha=0.5, label='Original')
plt.xlabel('Total Bedrooms')
plt.legend()

plt.subplot(2, 2, 4)
plt.hist(df['standardized_total_bedrooms'], bins=50, alpha=0.5, label='Standardized')
plt.xlabel('Standardized Total Bedrooms')
plt.legend()
```

```python
# Box plots
plt.figure(figsize=(12, 6))

plt.subplot(1, 4, 1)
plt.boxplot([df['total_rooms']], labels=['Original'])
plt.title('Total Rooms (Original)')

plt.subplot(1, 4, 2)
plt.boxplot([df['standardized_total_rooms']], labels=['Standardized'])
plt.title('Total Rooms (Standardized)')

plt.subplot(1, 4, 3)
plt.boxplot([df['total_bedrooms']], labels=['Original'])
plt.title('Total Bedrooms (Original)')

plt.subplot(1, 4, 4)
plt.boxplot([df['standardized_total_bedrooms']], labels=['Standardized'])
plt.title('Total Bedrooms (Standardized)')

plt.tight_layout()
plt.show()
```

# Question 3

```python
# Define predictor variable names
predictor_vars = ['housing_median_age', 'standardized_total_rooms_by_population', 'standardized_total_bedrooms_by_population', 
                  'population', 'households', 'median_income', 'ocean_proximity']

r_squared_values = []

# Create subplots
fig, axes = plt.subplots(3, 3, figsize=(16, 8))

# Flatten the axes for easier iteration
axes = axes.flatten()

# Loop through each predictor variable
for i, var in enumerate(predictor_vars):
    # Prepare the data
    X = df[var].values.reshape(-1, 1)
    y = df['median_house_value']
    
    # Fit linear regression model
    model = LinearRegression()
    model.fit(X, y)
    
    # Predict y values
    y_pred = model.predict(X)
    
    # Plot scatter plot and regression line
    ax = axes[i]
    ax.scatter(X, y, color='blue', alpha=0.5, label='Data')
    ax.plot(X, y_pred, color='red', label='Regression Line')
    ax.set_xlabel(var)
    ax.set_ylabel('median_house_value')
    ax.legend()
    
    # Calculate R-squared
    r_squared = r2_score(y, y_pred)
    r_squared_values.append(r_squared)
    ax.set_title(f'{var}')

plt.tight_layout()
plt.show()
```

```python
# Print R-squared values
for i, var in enumerate(predictor_vars):
    print(f"R-squared value for {var}: {r_squared_values[i]}")
```

# Question 4

```python
# Fit multiple linear regression model
X = sm.add_constant(df[predictor_vars]) 
multiple_model = sm.OLS(y, X).fit()

# Create subplots
fig, axes = plt.subplots(3, 3, figsize=(16, 8))

# Residuals vs. Fitted Values Plot
ax = axes[0, 0]
ax.scatter(multiple_model.fittedvalues, multiple_model.resid, alpha=0.5)
ax.set_xlabel("Fitted Values")
ax.set_ylabel("Residuals")
ax.set_title("Residuals vs. Fitted Values Plot")
ax.axhline(y=0, color='red', linestyle='--')

# Residuals vs. Predictor Plots
for i, var in enumerate(predictor_vars):
    row = (i + 1) // 3
    col = (i + 1) % 3
    ax = axes[row, col]
    ax.scatter(df[var], multiple_model.resid, alpha=0.5)
    ax.set_xlabel(var)
    ax.set_ylabel("Residuals")
    ax.axhline(y=0, color='red', linestyle='--')

plt.tight_layout()
plt.show()
```

```python
# Prepare the data
X = df[predictor_vars]
y = df['median_house_value']

# Fit multiple linear regression model
multiple_model = LinearRegression()
multiple_model.fit(X, y)

# Predict y values
y_pred_multiple = multiple_model.predict(X)

# Calculate R-squared for multiple regression model
r_squared_multiple = r2_score(y, y_pred_multiple)
print("R-squared for multiple regression model:", r_squared_multiple)

# Compare R-squared values with the individual linear regression models
print("R-squared values for individual predictor variables:")
for i, var in enumerate(predictor_vars):
    print(f"{var}: {r_squared_values[i]}")
```
